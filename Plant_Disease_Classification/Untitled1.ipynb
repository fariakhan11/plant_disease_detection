```python
import mlflow
import mlflow.keras
from tensorflow.keras.models import load_model
import tensorflow as tf
from tensorflow.keras.applications import Xception

# Rebuild architecture similar to the saved model
model = Xception(weights=None, include_top=True)

# Load only the weights (ignore mismatched layer names)
model.load_weights("Xception.h5", by_name=True, skip_mismatch=True)

print("✅ Model and weights loaded successfully!")

# 2️⃣ Set up local MLflow tracking
mlflow.set_tracking_uri("sqlite:///mlflow.db")  # stores tracking info in a local SQLite DB
mlflow.set_experiment("Xception_Deployment")

# 3️⃣ Log the model + basic info
with mlflow.start_run(run_name="best_xception_model"):
    mlflow.keras.log_model(model, "model")
    mlflow.log_metric("accuracy", 0.99)
    mlflow.log_param("architecture", "Xception")
    mlflow.log_param("dataset_split", "train/val/test")
```
