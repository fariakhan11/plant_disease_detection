```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# ✅ Ensure input size matches Xception model requirements
input_shape = (299, 299)  # Changed from (228, 228) to (299, 299)

# Load the trained Xception model
model_path = "Xception.h5"  # Ensure this model file exists
model = tf.keras.models.load_model(model_path)

# Define actual class labels (in correct order)
true_class_labels = [
    'Corn Common Rust', 'Corn Gray Leaf Spot', 'Corn Healthy', 'Corn Northern Leaf Blight',
    'Potato Early Blight', 'Potato Healthy', 'Potato Late Blight',
    'Rice Brown Spot', 'Rice Healthy', 'Rice Leaf Blast', 'Rice Neck Blast',
    'Wheat Brown Rust', 'Wheat Healthy', 'Wheat Yellow Rust'
]

# Fetch correct class labels dynamically to avoid mismatches
train_dir = "C:\\Dataset\\Train\\"  # Replace with actual dataset path
datagen_train = ImageDataGenerator(rescale=1./255, dtype=np.float32)
generator_train = datagen_train.flow_from_directory(directory=train_dir, 
                                                    target_size=input_shape, 
                                                    batch_size=1, 
                                                    shuffle=False)

# Get assigned class indices from flow_from_directory()
class_indices = generator_train.class_indices  # Example: {'0': 0, '1': 1, ..., '13': 13}

# Correct mapping: Model index → Actual disease label
index_to_class = {v: true_class_labels[int(k)] for k, v in class_indices.items()}

# Possible causes of diseases
disease_causes = {
    'Corn Common Rust': "Caused by the fungus *Puccinia sorghi*. It spreads in humid conditions.",
    'Corn Gray Leaf Spot': "Caused by *Cercospora zeae-maydis*. Occurs due to warm temperatures and humidity.",
    'Corn Healthy': "No disease detected. The leaf is healthy.",
    'Corn Northern Leaf Blight': "Caused by *Exserohilum turcicum*. Common in wet and cool conditions.",
    'Potato Early Blight': "Caused by *Alternaria solani*. Manifests as dark spots on older leaves.",
    'Potato Healthy': "No disease detected. The leaf is healthy.",
    'Potato Late Blight': "Caused by *Phytophthora infestans*. Spreads in wet, humid weather.",
    'Rice Brown Spot': "Caused by *Cochliobolus miyabeanus*. Often due to nutrient deficiency.",
    'Rice Healthy': "No disease detected. The leaf is healthy.",
    'Rice Leaf Blast': "Caused by *Magnaporthe oryzae*. Thrives in warm, humid environments.",
    'Rice Neck Blast': "A more severe form of Rice Blast affecting the neck region of the plant.",
    'Wheat Brown Rust': "Caused by *Puccinia triticina*. Favors cool, moist conditions.",
    'Wheat Healthy': "No disease detected. The leaf is healthy.",
    'Wheat Yellow Rust': "Caused by *Puccinia striiformis*. Common in cool, wet weather."
}

# Define remedies for each disease
disease_remedies = {
    'Corn Common Rust': "Apply fungicides containing azoxystrobin or mancozeb. Rotate crops and plant resistant varieties.",
    'Corn Gray Leaf Spot': "Use resistant hybrids. Apply foliar fungicides like pyraclostrobin. Reduce field humidity.",
    'Corn Healthy': "No action needed. Maintain good farming practices.",
    'Corn Northern Leaf Blight': "Apply fungicides containing propiconazole. Ensure proper crop rotation.",
    'Potato Early Blight': "Use copper-based fungicides. Improve field drainage and remove infected leaves.",
    'Potato Healthy': "No action needed. Maintain optimal soil health.",
    'Potato Late Blight': "Use fungicides like chlorothalonil. Avoid overhead irrigation and use resistant varieties.",
    'Rice Brown Spot': "Apply potassium-rich fertilizers. Improve soil nutrition and water management.",
    'Rice Healthy': "No action needed. Continue good farming practices.",
    'Rice Leaf Blast': "Use blast-resistant varieties. Apply tricyclazole-based fungicides and avoid excessive nitrogen use.",
    'Rice Neck Blast': "Improve field drainage. Use disease-resistant seeds and apply silicon-based fertilizers.",
    'Wheat Brown Rust': "Apply fungicides like tebuconazole. Remove and destroy infected plant debris.",
    'Wheat Healthy': "No action needed. Maintain good farming practices.",
    'Wheat Yellow Rust': "Use rust-resistant varieties. Apply fungicides like triazoles before infection spreads."
}

# Function to preprocess a single image
def preprocess_image(image_path):
    """Loads and preprocesses a single image for classification."""
    img = load_img(image_path, target_size=input_shape)  # ✅ Resize to 299x299
    img_array = img_to_array(img) / 255.0  # Normalize pixel values
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    return img_array, img

# Function to classify a single leaf image
def predict_leaf_disease(image_path):
    """Classifies the given image and displays the result."""
    img_array, img = preprocess_image(image_path)

    # Predict using the trained model
    predictions = model.predict(img_array)
    predicted_class_index = np.argmax(predictions)  # Get highest probability class
    predicted_class = index_to_class[predicted_class_index]  # Convert index to class label
    confidence = np.max(predictions) * 100  # Confidence percentage

    # Display image with prediction
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"Predicted: {predicted_class} ({confidence:.2f}%)", fontsize=14, color="green")
    plt.show()

    # Print results
    print(f"Predicted Class: {predicted_class}")
    print(f"Confidence: {confidence:.2f}%")
    print(f"Possible Cause: {disease_causes[predicted_class]}")
    print(f"Remedies: {disease_remedies[predicted_class]}")

# Example usage
if __name__ == "__main__":
    image_path = r"C:\Dataset\Test\4\0e0a1b51-f61c-4934-bc57-a820af1faacb___RS_Early.B 7147.jpg"  # Replace with actual image path
    predict_leaf_disease(image_path)
```

```python
import tensorflow as tf

# Load your trained Keras model
model = tf.keras.models.load_model("Xception.h5")

# Convert the model to TensorFlow Lite format
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

# Save the converted model
with open("Xception.tflite", "wb") as f:
    f.write(tflite_model)

print("Model successfully converted to TensorFlow Lite!")
```
