```python
# =========================================
# üåø Plant Disease Detection - Xception Evaluation with MLflow
# =========================================

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import Xception
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Flatten, Dropout, Dense
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, classification_report
import numpy as np
import mlflow
import mlflow.keras

# =========================================
# 1Ô∏è‚É£ Rebuild architecture (same as training)
# =========================================
input_shape = (299, 299, 3)

base_model = Xception(include_top=False, input_shape=input_shape, weights='imagenet')
conv_model = Model(inputs=base_model.input, outputs=base_model.output)

model = Sequential([
    conv_model,
    Flatten(),
    Dropout(0.5),
    Dense(512, activation='relu'),
    Dense(14, activation='softmax')
])

# Compile for evaluation
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Load the trained weights
model.load_weights("Xception.h5")
print("‚úÖ Model architecture rebuilt and weights loaded successfully!")

# =========================================
# 2Ô∏è‚É£ Load Test Dataset
# =========================================

test_dir = "C:\\Dataset\\Test\\"
input_size = (299, 299)
batch_size = 32

datagen_test = ImageDataGenerator(rescale=1./255)

generator_test = datagen_test.flow_from_directory(
    directory=test_dir,          # same directory path used during training
    target_size=input_size,
    batch_size=batch_size,
    shuffle=False
)

# =========================================
# 3Ô∏è‚É£ Evaluate Model
# =========================================
Y_pred = model.predict(generator_test)
y_pred = np.argmax(Y_pred, axis=1)
y_true = generator_test.classes

accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='weighted', zero_division=0)
recall = recall_score(y_true, y_pred, average='weighted', zero_division=0)
f1 = f1_score(y_true, y_pred, average='weighted', zero_division=0)

print("\n‚úÖ Evaluation Metrics:")
print(f"Accuracy : {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"F1 Score : {f1:.4f}")

# Optional detailed report
class_labels = list(generator_test.class_indices.keys())
print("\nüìä Classification Report:\n", classification_report(y_true, y_pred, target_names=class_labels))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
import matplotlib.pyplot as plt

# y_true and y_pred are your true/predicted labels
cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues')

plt.savefig("confusion_matrix.png")
mlflow.log_artifact("confusion_matrix.png")

# Log classification report
report = classification_report(y_true, y_pred, output_dict=True)
for label, metrics in report.items():
    if isinstance(metrics, dict):
        for metric_name, value in metrics.items():
            mlflow.log_metric(f"{label}_{metric_name}", value)
```
